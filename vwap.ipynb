```python
import yfinance as yf
import pandas as pd
from ta.volume import VolumeWeightedAveragePrice
```

```python
def vwapfn(dataframe, label='vwap', window=3, fillna=True):
        dataframe[label] = VolumeWeightedAveragePrice(high=dataframe['High'], low=dataframe['Low'], close=dataframe["Close"], volume=dataframe['Volume'], window=window, fillna=fillna).volume_weighted_average_price()
        return dataframe
```

```python
df = yf.Ticker("AAPL").history(period="1d", interval="5m")
df1 = vwapfn(df)
```

```python
df= df[df['Volume'] != 0]
df["ap"] = (df["High"]+df["Low"]+df["Low"])/3
df["wap"] = df["ap"]* df["Volume"]
```

```python
df =df.drop(columns ={"Open", "High", "Low", "Close", "Dividends","Stock Splits" })
df
```

```python
df["vwap"]= df["wap"].rolling(5).sum()/ df["Volume"].rolling(5).sum()
```
